## Project Title: Cultural Story Generator
### Project Concept: the goal is to design a creative Large Language Model Application 
The idea is to create a project that will have lots of fun and interactive. Imagine visiting a museum or art gallery and being able to ask a tool for the story behind a painting. Not just any story, but one that can change based on your mood. Want the story to be dramatic? Or maybe you’d like it to be funny? What if you want the explanation in Gen Z slang? This tool will use a large language model (LLM) to generate creative, engaging stories behind artworks, with the bonus of music to set the scene
### Features:
We will go throught the main features, let's start with the first one, the users can choose an artwork from a collection in a museum or gallery, and  then ask the app, which will be designed as a chatbox, for a detailed story behind it has. The LLM will tell a story about the painting, and will share the details like who is the artist, maybe their biography, when the painting was created, and what the painting might symbolise or represent from that time period.
One of the key feature is that the user can customise the story. For example,  if the user wants to hear the story in a dramatic, tragic tone, the LLM will do this. The app will choose what the story should be described based on the story behind the painting. More making the app more engaged the user can ask the tool to explain the painting using modern slang, like how a Gen Z nowadays talk
For making the app more immersive, the app can accompany the story with music that fits the vibe of the painting or fits to the feelings of the painting. For example, classical music will be played for a Renaissance painting period, whereas something more upbeat can be played with contemporary art
Instead of just reading about a painting, the user can engage with it by asking some questions, stories, and play around with different styles of narration. It should be not boring as if you will sit in the lecture room and listen to a professor, it’s an interactive experience that makes learning about culture and history fun
 #### Why it's a good idea
Most of the time people find museums or art galleries boring because they don't understand the painting and how deep the story it may have. So instead of trying to guess what the specific art means, the user can ask its story and learn about it how it suits him. Plus, with the added humour and slang features, it’s a way to make art more accessible, especially to younger audiences who usually finds traditional old museum tours boring. So this is good for user engagement, it can be personalised based on user's taste and it's good for cultural learning 
#### User Experience and Accessibility:
The interface of the app should be easy to use where users can interact with the paintings, even by connecting the same feature as in ChatGPT-4 where you can take a picture of the object and explain the painting by selecting their preferred storytelling mode. It can be accessed through both web and mobile platforms to ensure widespread accessibility, but if the user plans to take a picture it will not be convenient to do it through the laptop, and also the user should mind the museum's or art gallery rules because some of them has restrictions of taking a picture
The app will have an access to all languages, allowing users from different cultural backgrounds to engage with artworks in their native language. That will help to increase usage because more people will understand how to use the app
Users could engage with the artwork beyond just reading a story. For example, they could ask the system to “describe the artist’s emotions while creating this” or “what would the artist think of modern interpretations of this work.” These types of interactions would make the experience more immersive and personal.
### Main audience
Since we have an interactive features that app will be good for **young learners** because they are often more comfortable with technology and interactive learning experiences. The ability to customise the tone of stories and include modern elements like slang or humor could resonate well with Gen Z
Also it will be good for general **visitors**. rather than relying on traditional static plaques or guided tours, visitors could use the app to explore deeper meanings, ask questions, and receive personalised insights into the works they’re viewing in real time

#### About technology 
The app will be basically built on Language Model (LLM) features it will rely on a pre-trained large language model, such as GPT-3 or GPT-4, which would generate text based on the input. These models can be accessed through APIs provided by OpenAI, or if you're using a local model, libraries like Hugging Face Transformers can be employed to load the model. But the problem is that the user will have limited about of prompts, otherwise he would need to pay for the service, since the project will by API keys for accessing more prompts 


Also in terms of musical accompaniment we would need  Spotify API to provide music recommendations, based on the painting and users requests.

User Interaction: The app would need a text input or dropdown for the user to customise the story, so it should be something simple and most-likely similar to ChatGPT interface 


#### Integration with Large Language Models (LLMs):
Firstly we would need to do the data preprocessing, to allow the LLM to understand and generate appropriate stories. For instance, descriptions of the artwork, artist background, and historical context will be fed into the model, and it will use its training to craft a personalised narrative. 
Based on user preferences, such as the tone of the story or the use of slang, different prompts could be passed to the model. For example, the user can have some similar prompts "Tell me the story of this painting in a dramatic, tragic tone.", "Explain this artwork using modern Gen Z slang."
Music would be played alongside the generated text, possibly synced with key points of the story to enhance the immersive experience. Depending on the painting's theme, music genres could be matched, it will be done as mentioned above by connecting Spotify's APIs


#### The technical and user workflow
User writes the name of the painting and the author to the chatbox, then he should provide the specific prompt and maybe ask for musical accompaniment. The app eats the prompt and gives you appropriate output
